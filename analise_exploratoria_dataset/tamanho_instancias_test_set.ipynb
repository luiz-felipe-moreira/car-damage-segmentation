```python
import json
import pandas as pd
```

```python
test_labels_path = "/home/s825860045/Datasets/CarDD_release/CarDD_COCO/annotations/instances_test2017.json"
```

```python
with open(test_labels_path,"r") as file:
    jsonData = json.load(file)
 
print("Type of JSON Object: ", type(jsonData))
```

```python
annotations = jsonData["annotations"]
```

```python
annotations
```

```python
annotations_list = []
for annotation in annotations:
    data = [annotation["id"], annotation["image_id"], annotation["category_id"], annotation["area"]]
    annotations_list.append(data)
annotations_list
```

```python
data = pd.DataFrame(annotations_list, columns=["id", "image_id", "category_id", "area"])
data
```

```python
categories = {
    1: "dent",
    2: "scratch",
    3: "crack",
    4: "glass shatter",
    5: "lamp broken",
    6: "tire flat"
}
data["category_name"] = data["category_id"].map(categories)
data
```

```python
counts_all_sizes = data["category_name"].value_counts()
counts_all_sizes
```

```python
small_areas = data[data["area"] < 32 ** 2]
small_areas
```

```python
counts_small_size = small_areas["category_name"].value_counts()
counts_small_size
```

```python
medium_areas = data[(data["area"] > 32 ** 2) & (data["area"] < 96 ** 2)]
medium_areas
```

```python
counts_medium_size = medium_areas["category_name"].value_counts()
counts_medium_size
```

```python
large_areas = data[data["area"] > 96 ** 2]
large_areas
```

```python
counts_large_size = large_areas["category_name"].value_counts()
counts_large_size
```
